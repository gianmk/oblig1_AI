```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/umaimehm/Intro_to_AI/master/assignment1/Ruter_data.csv'
data = pd.read_csv(url, delimiter = ';')
```

```python
data.head()
```

```python
data.shape
```

```python
plt.hist(data.Passasjerer_Ombord, bins=100, rwidth=0.5)
plt.xlabel('passasjer (Antall)')
plt.ylabel('Count')
plt.show()
```

```python
#Identifies some high outliers
threshold_max = data['Passasjerer_Ombord'].quantile(0.999)
threshold_max
```

```python
data[data['Passasjerer_Ombord']>threshold_max]
```

```python
#Identifies some love outliers
threshold_min = data['Passasjerer_Ombord'].quantile(0.001)
threshold_min
```

```python
data[data['Passasjerer_Ombord']<threshold_min]
```

```python
new_data=data[(data['Passasjerer_Ombord']<threshold_max) & (data['Passasjerer_Ombord']>threshold_min)]
```

```python
new_data.shape
```

```python
#Make new feature Måned
data['Måned'] = pd.to_datetime(data.Dato, dayfirst=True).dt.month_name()
```

```python
#Histogram showing number of registrations in each month
plt.hist(data.Måned, bins = 12, rwidth = 10)
plt.xlabel('Måned')
plt.ylabel('Number of buses going')

plt.show()
```

```python
data.head()
```

```python
# feature to see busses going to skallum at specific date.
data[(data.Dato=="03/08/2020")&(data.Holdeplass_Til=="Skallum")]
```

```python
#feature to see Linjetype at specific date. 
data[(data.Dato=="27/07/2020")&(data.Linjetype=="Lokal")]
```

```python
#feature to see total amount of bus rides in august. 
data[(data.Måned=="August")&(data.TurId)]
```

```python
#Visualization
```

```python
plt.figure(figsize=(12,8))
data.loc[data.Kommune == 'Asker'].groupby('Linjenavn').TurId.count().plot(kind = 'bar',)

plt.ylabel('amount of buses')
```

```python
plt.figure(figsize=(12,8))
data.groupby('Område').Kjøretøy_Kapasitet.mean().plot(kind = 'bar', color=plt.cm.Paired(np.arange(len(data.Område))))
plt.ylabel('Kapasitet')
plt.title('Gjennomsnittkapasitet in forskjellig område', fontsize=15)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='Fylke', y='Kjøretøy_Kapasitet',hue='Linjeretning',data=data)
```

```python
#The change in passangers on the stop, and compares it to the total capacity on the bus. Gives out a number in percent
data['Prosent_Passasjerendring_Pr_Kapasitet'] = (data['Passasjerer_Ombord']/data['Kjøretøy_Kapasitet'])*100
```

```python
data.sort_values('Prosent_Passasjerendring_Pr_Kapasitet')
```

```python
#Diagram showing the mean change in passangers on a bus stop in the different municipalities
plt.figure(figsize=(12,8))
data.groupby('Kommune').Prosent_Passasjerendring_Pr_Kapasitet.mean().plot(kind = 'bar', color=plt.cm.Paired(np.arange(len(data.Kommune))))
plt.ylabel('Prosent')
plt.title('Gjennomsnittlig prosentandel påstigninger iforhold til avstigninger pr kommune', fontsize=15)
```
